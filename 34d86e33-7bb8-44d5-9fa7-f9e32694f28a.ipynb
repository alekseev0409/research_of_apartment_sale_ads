### Добавим в таблицу новые столбцы

**Посчитаем значения и добавим следующие столбцы в таблицу:**
- price_one_square_meter: цена одного квадратного метра(нужно поделить стоимость объекта на его общую площадь, а затем округлить до двух знаков после запятой);
- exposition_weekday: день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
- exposition_month: месяц публикации объявления;
- exposition_year: год публикации объявления;
- floor_category: тип этажа квартиры (значения — «первый», «последний», «другой»);
- city_centers_nearest_km: расстояние до центра города в километрах (переведите из м в км и округлите до целых значений).

```python
# Добавим столбец с ценой за 1м² и округлим до двух знаков после запятой
data['price_one_square_meter'] = (data['last_price'] / data['total_area']).round(2)
```

```python
data.head(10)
```

**Добавлю столбец с неделей, месяцем и годом публикации:**

```python
# Добавляю столбец с днем недели
data['exposition_weekday'] = data['first_day_exposition'].dt.weekday

# Добавляю столбец с месяцем 
data['exposition_month'] = data['first_day_exposition'].dt.month

# Добавляю столбец с годом
data['exposition_year'] = data['first_day_exposition'].dt.year
```

```python
data.head(5)
```

**Добавлю столбец с категорией этажа:**

```python
def get_floor_category(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
    
data['floor_category'] = data.apply(get_floor_category, axis=1)
```
